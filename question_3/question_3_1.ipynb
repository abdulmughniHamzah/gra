**1. Generate a matrix of charts with 3 rows and 4 columns containing the histograms of all the data columns (except ID and quality)**

***Answer***



Code below generates histograms representing the data distribution for all 11 measurements provided

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
################################################################################################################
# Read data
data = pd.read_csv("../data/question_3.csv")
################################################################################################################
# 3x4 grid of histograms
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
fig.suptitle('Histograms')
# drop ID, and Quality
data = data.drop(columns=['quality', 'Id'], axis=1)
columns = data.columns
for i in range(3):
    for j in range(4):
        column_indix = i * 4 + j
        if column_indix < len(columns):
            axes[i, j].hist(data[columns[column_indix]], bins=10, color='skyblue')
            axes[i, j].set_title(columns[column_indix])
            axes[i, j].set_xlabel("Value")
            axes[i, j].set_ylabel("Frequency")
# Adjust the layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
# Display the plot
plt.savefig('../output/question_3_1_histograms.png', bbox_inches='tight')
```
