**3. Calculate the Pearson’s correlation coefficient for all the combination of variables in the dataset (except ID) and visualize it as cross matrix**

***Answer***

The following code calculates the correlation matrix using the corr function available in the DataFrame and visualizes it as a heatmap.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
################################################################################################################
# Read data
data = pd.read_csv("../data/question_3.csv")
################################################################################################################
# exclude Id
data = data.drop(columns=['Id'], axis=1)
# Pearson's correlation coefficients
correlations = data.corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Pearson's Correlation Coefficient Matrix")
plt.savefig('../output/question_3_3_correlations.png', bbox_inches='tight')
```
