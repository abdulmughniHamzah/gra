**5. Is it possible to project the test variation for future orders 7 and 8 based on the given data? Why or why not? Is there some additional information that will be needed for this purpose?**

***Answer***

In order to answer this question, a statistical time series analysis is performed. The analysis employs "Unit No" as a time surrogate, assuming that measurements within the same "Lot No" are taken sequentially. So, the python code below generate the line chart by product code to visualize the measurement patterns. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
################################################################################################################
measurements = pd.read_excel('../data/question_1_2.xlsx', sheet_name='Measurements', skiprows=1)
specifications = pd.read_excel('../data/question_1_2.xlsx', sheet_name='Specifications', skiprows=1)
################################################################################################################
measurements.set_index([ 'Unit_No'], inplace=True)

# Plot T_Height and T_Weight time series
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('T_Height Time Series')
plt.xlabel('Unit No')
plt.ylabel('T_Height')
for product_code, group in measurements.groupby('Product Code'):
    plt.plot(group.index, group['T_Height'], marker='o', label=product_code)
plt.legend()

plt.subplot(1, 2, 2)
plt.title('T_Weight Time Series')
plt.xlabel('Unit No')
plt.ylabel('T_Weight')
for product_code, group in measurements.groupby('Product Code'):
    plt.plot(group.index, group['T_Weight'], marker='o', label=product_code)
plt.legend()

plt.tight_layout()
plt.savefig('../output/question_1_5_time_series.png', bbox_inches='tight')

```

**Referring to the time series data generated by the previous code, I can highlights these key points**

- **Weight Projection**: The ***T_Weight*** data shows a consistent pattern, suggesting it could be projected. However, the dataset's small size emphasizes the need for obtaining more data points to enhance projection confidence.

- **Height Projection**: The ***T_Height*** projection is challenging due to the absence of a clear and predictable trend in the current dataset.

- The analysis recognizes the limitations of a small dataset and the presence of outliers, both of which can undermine the feasibility of making future projections. Small datasets may result in less reliable predictions, and anomalies can disrupt patterns, further reducing projection reliability.
